#### Feature Engineering

```python
df['Date'] = df['Date_of_Journey'].str.split('/').str[0]
df['Month'] = df['Date_of_Journey'].str.split('/').str[1]
df['Year'] = df['Date_of_Journey'].str.split('/').str[2]
```

```python
df.head()
```

Here, our newly added columns are of datatype *object* as well. Now, we will convert them to *int* datatype.

```python
df['Date'] = df['Date'].astype(int)
df['Month'] = df['Month'].astype(int)
df['Year'] = df['Year'].astype(int)
```

```python
df.info()
```

Drop *Date_of_Journey* because there is no need at all

```python
df.drop('Date_of_Journey', axis=1, inplace=True)
df.head()
```

For the *Arrival_Time*, we will get the *hours* and *minutes* are splitted.<br>
We need to do some coding and other logical thinking to get the appropriate output.

```python
# First get rid of the extra date given in the Arrival_Time
df['Arrival_Time'] = df['Arrival_Time'].apply(lambda x : x.split(' ')[0])

# New, we can separate the hours and minutes
df['Arrival_hour'] = df['Arrival_Time'].str.split(':').str[0]
df['Arrival_minute'] = df['Arrival_Time'].str.split(':').str[1]
```

```python
df.head(2)
```

Again, we will convert the datatype to *int* for the newly created column

```python
df['Arrival_hour'] = df['Arrival_hour'].astype(int)
df['Arrival_minute'] = df['Arrival_minute'].astype(int)
df.info()
```

Now, we will drop *Arrival_Time* because it of no use now

```python
df.drop('Arrival_Time', axis=1, inplace=True)
df.head(2)
```

Now, we will do the same with *Dep_Time* as well

```python
df['Departure_hour'] = df['Dep_Time'].str.split(':').str[0]
df['Departure_minute'] = df['Dep_Time'].str.split(':').str[1]
df.head(2)
```

Again, we will convert the datatype of newly created column to *int*

```python
df['Departure_hour'] = df['Departure_hour'].astype(int)
df['Departure_minute'] = df['Departure_minute'].astype(int)
df.info()
```

Now, we will drop *Dep_Time* because it is no longer needed at all

```python
df.drop('Dep_Time', axis=1, inplace=True)
df.head(2)
```

#### Categorical Features

```python
df['Total_Stops'].unique()
```

Check for the data where *Total_Stops* is *nan*

```python
df[df['Total_Stops'].isnull()]
```

Now, we have to replace the *nan* value with the <strong>mode</strong> of the column

```python
df['Total_Stops'].mode()
```

Since the <strong>mode</strong> is *'1 stop'*, therefore we will fill the *nan* with that.

```python
df['Total_Stops'] = df['Total_Stops'].map({
     'non-stop':0, '1 stop':1, '2 stops':2, '3 stops':3, '4 stops':4, np.nan:1
})
```

Let's confirm that there is no flight with *nan* stops

```python
df[df['Total_Stops'].isnull()]
```

```python
df.head(2)
```

We can drop the *Route* column because we already have *Source* and *Destination*

```python
df.drop('Route', axis=1, inplace=True)
df.head(2)
```

Let's handle the *Airline* column

```python
df['Airline'].unique()
```

```python
df['Source'].unique()
```

```python
df['Destination'].unique()
```

We will use *OneHotEncoder* to convert these categorical variables to *int* format

```python
from sklearn.preprocessing import OneHotEncoder
encoder = OneHotEncoder()
```

Now, we will use the *OneHotEncoder* to fit these features.

```python
encoder.fit_transform(df[['Airline','Source','Destination']]).toarray()
```

```python
pd.DataFrame(encoder.fit_transform(df[['Airline','Source','Destination']]).toarray(), columns=encoder.get_feature_names_out())
```
