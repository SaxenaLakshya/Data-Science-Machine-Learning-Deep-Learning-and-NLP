## Label Encoding

```python
import pandas as pd

df = pd.DataFrame({
    'color': ['red', 'blue', 'green', 'green', 'red', 'blue']
})
```

```python
df.head()
```

Importing *Label Encoder* from the library.

```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
```

```python
label_encoder.fit_transform(df[['color']])
```

```python
label_encoder.transform([['red']])
```

```python
label_encoder.transform([['blue']])
```

```python
label_encoder.transform([['green']])
```

In case of assigning *Rank* in the models, our model might think that *'red'* is higher than *'blue'* and *'green'*. So on and so forth...<br>
In these kinds of models, we use <strong>*Ordinal Encoding*</strong>

## Ordinal Encoding

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
# Create a sample dataframe with an Ordinal variable
df = pd.DataFrame({
    'size': ['small', 'medium', 'large', 'medium', 'small', 'large']
})
```

```python
# Create an instance of OrdinalEncoder and then fit_transform
ordinal_encoder = OrdinalEncoder(categories=[['small','medium','large']])
ordinal_encoder.fit_transform(df[['size']])
```

Here, we are providing categories the number/rank intentionally. Using this, we can handle the model with ranks correctly.<br>
We might use this encoder where our model needs context of ranks and positions.
