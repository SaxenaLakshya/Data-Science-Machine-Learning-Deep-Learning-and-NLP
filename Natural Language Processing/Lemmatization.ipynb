# WordNet Lemmatization
NLTK provides `WordNetLemmatizer` class which is a thin wrapper around the wordnet corpus. This class uses morphy() function to the WordNet CorpusReader class to find a lemma. Let us understand it with an example.

```python
from nltk.stem import WordNetLemmatizer

lemmatizer = WordNetLemmatizer()
```

In the `WordNetLemmatizer` class, we have one special parameter i.e. `pos`. It basically, differentiate the kind of the word given.<br>
The following are the kinds supported by this class:<br>
- Noun : n (default)
- Verb : v
- Adjective : a
- Adverb : r

```python
import nltk

nltk.download('wordnet')
nltk.download('omw-1.4')
```

The above cell must be run in order to not to get the errors. To run the Lemmatization, `wordnet` is required which is not available by default.

```python
lemmatizer.lemmatize("going", pos="v")
```

```python
words = ["eat", "eats", "eaten", "writing", "writes", "programming", "programs", "history", "finally", "finalize"]
for word in words:
    print(word + " --> " + lemmatizer.lemmatize(word, pos="v"))
```

```python
lemmatizer.lemmatize("fairly"), lemmatizer.lemmatize("sportingly")
```

*Summary*: **Lemmatization** will take more time as compared to the **Stemming** because it compare from the `wordnet` in order to get the accurate answer.<br>
It can be used in Q&A, ChatBots and Text Summarization.
