# Ridge, Lasso and ElasticNet Regression
*We are creating a project based on Algerian Forest Fires*<br>
Importing Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

Reading the dataset<br>
Since, the first line in CSV has the title, we need to remove it while reading it using *header=1*

```python
dataset = pd.read_csv("Algerian_forest_fires_dataset_UPDATE.csv", header=1)
dataset.head()
```

Getting the general information about the dataset

```python
dataset.info()
```

## Data Cleaning
Check for missing values

```python
dataset[dataset.isnull().any(axis=1)]
```

The dataset is converted into two sets based on Region from 122th index, we make a new column on the Region.<br>
1 : "*Bejaia Region Dataset*"<br>
2 : "*Sidi-Bel Abbes Region Dataset*"<br>
Add new column with region

```python
dataset.loc[:122, "Region"] = 0
dataset.loc[122:, "Region"] = 1
df = dataset
df.info()
```

Convert the value of the *Region* to *int* to decrease the space

```python
df[['Region']] = df[['Region']].astype(int)
df.head()
```

Removing the *NULL* values

```python
df = df.dropna().reset_index(drop=True)
df.isnull().sum()
```

Remove the 122th row as well because it is a repeated row containing column names only

```python
df = df.drop(122).reset_index(drop=True)
df.iloc[[122]]
```

Fix spaces in columns names

```python
df.columns = df.columns.str.strip()
df.columns
```

Change the required columns as integer data type

```python
df[['day', 'month', 'year', 'Temperature', 'RH', 'Ws']] = df[['day', 'month', 'year', 'Temperature', 'RH', 'Ws']].astype(int)
df.info()
```

Change the required columns as float data type

```python
df[['Rain', 'FFMC', 'DMC', 'DC', 'ISI', 'BUI', 'FWI']] = df[['Rain', 'FFMC', 'DMC', 'DC', 'ISI', 'BUI', 'FWI']].astype(float)
df.info()
```

Finally, we can describe our dataset

```python
df.describe()
```

Let's save the cleaned dataset

```python
df.to_csv('Algerian_forest_fires_dataset_cleaned.csv', index=False)
```

## Exploratory Data Analysis
We will drop the *day, month and year*

```python
df_copy = df.drop(['day', 'month', 'year'], axis=1)
df_copy.head()
```

Convert *Classes* to 0's and 1's

```python
df_copy['Classes'] = np.where(df_copy['Classes'].str.contains('not fire'), 0, 1)
df_copy['Classes'].value_counts()
```

Plot density plot for all features

```python
df_copy.hist(bins=50, figsize=(20, 15))
plt.show()
```

Percentage for Pie Chart

```python
percentage = df_copy['Classes'].value_counts(normalize=True) * 100
```

Plotting the Pie Chart

```python
classlabels = ['Fire', 'Not Fire']
plt.figure(figsize=(12, 7))
plt.pie(percentage, labels=classlabels, autopct='%1.1f%%')
plt.title('Pie Chart of classes')
plt.show()
```

Correlation

```python
df_copy.corr()
```

Box Plots to check for the *Outliers*

```python
sns.boxplot(df_copy['FWI'], color='orange')
```

Monthly Fire Analysis (Bejaia Region)

```python
# Do the changes in df iteself
df['Classes'] = np.where(df['Classes'].str.contains('not fire'), 'No Fire', 'Fire')

df_temp = df.loc[df['Region'] == 1]
plt.subplots(figsize=(13, 6))
sns.set_style('whitegrid')
sns.countplot(x='month', hue='Classes', data=df)
plt.xlabel('Months', weight='bold')
plt.ylabel('Number of Fires', weight='bold')
plt.title('Fire Analysis of Bejaia Regions', weight='bold')
```
